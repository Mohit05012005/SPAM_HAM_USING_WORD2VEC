```python
## why we use average word2vec:
ls = []
def average_word2vec(doc):
  for line in doc:
    ls.append(np.mean([model.wv[word] for word in line if word in model.wv.index_to_key]))
  return ls
```

```python
from tqdm import tqdm

avg = []
for word_list in tqdm(words, desc="Calculating average word vectors"):
  vector_sum = np.zeros(model.wv['hello'].shape[0])
  count = 0
  for w in word_list:
    if w in model.wv.index_to_key:
      vector_sum = vector_sum + model.wv[w]
      count += 1
  if count > 0:
    avg.append(vector_sum/count)
  else:
    avg.append(np.zeros(model.wv['hello'].shape[0]))
```

```python
## steps to do before predicting it from the model.predict()
def cleaning_processing(string):
  review = re.sub('[^a-zA-Z]',' ',string)
  review = review.lower()
  review = review.split()
  review = [lemmatizer.lemmatize(word) for word in review if word not in stop_words]
  review = ' '.join(review)
  return review
```

```python
def vectorization(stringg):
  # Split the string into words and check if each word is in the model's vocabulary
  stringg = stringg.split()
  arr = [model.wv[word] for word in stringg if word in model.wv.index_to_key]
  return arr
```
